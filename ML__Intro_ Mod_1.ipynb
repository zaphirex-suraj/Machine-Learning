## Q1. Explain the differences between AI, ML, Deep Learning (DL), and Data Science (DS).

Answer-  Artificial Intelligence (AI), Machine Learning (ML), Deep Learning (DL), and Data Science (DS) are interconnected fields, yet they differ in scope, methods, and applications.  

### 1. Artificial Intelligence (AI)  
- **Definition:** AI is the broad science of making machines mimic human intelligence (reasoning, problem-solving, learning).  
- **Example:** Voice assistants like *Siri* or *Alexa*.  

### 2. Machine Learning (ML)  
- **Definition:** A subset of AI that focuses on building algorithms that improve automatically through experience.  
- **Approach:** Learns from **data patterns** rather than explicit programming.  
- **Example:** Email spam filters that improve with usage.  

### 3. Deep Learning (DL)  
- **Definition:** A subset of ML that uses **artificial neural networks** with multiple layers to model complex patterns.  
- **Use Case:** Works effectively with large datasets and high computational power.  
- **Example:** Facial recognition in smartphones.  

### 4. Data Science (DS)  
- **Definition:** A multidisciplinary field combining **statistics, ML, data analysis, and domain knowledge** to extract insights from data.  
- **Role:** Broader than ML—deals with data collection, cleaning, visualization, and decision-making.  
- **Example:** Analyzing customer purchase behavior to recommend products.  

---

### Key Distinction Table  

| Aspect | AI | ML | DL | DS |
|--------|----|----|----|----|
| Scope | Broad (human-like intelligence) | Subset of AI | Subset of ML | Broader, data-driven |
| Dependency | Algorithms | Data & patterns | Neural networks | Data + statistics |
| Example | Chess-playing bot | Spam filter | Self-driving cars | Business analytics |

---

### Conclusion  
AI is the overarching field, ML is a core method within AI, DL is a powerful specialized branch of ML, while DS is a broader practice of using data (with AI/ML as tools) to drive real-world decisions.  

## Q2. What are the types of machine learning? Describe each with one real-world example.


Answer- Machine Learning (ML) can be broadly categorized into three main types based on the nature of data and learning methods.  

---

### 1. Supervised Learning  
- **Definition:** The model learns from **labeled data**, i.e., data with input-output pairs.  
- **Goal:** Predict outcomes for new inputs by learning the mapping function.  
- **Techniques:** Regression, Classification.  
- **Example:**  
  - Predicting house prices using features like area, location, and number of rooms.  
  - Input → (features of the house), Output → (price).  

---

### 2. Unsupervised Learning  
- **Definition:** The model works on **unlabeled data**, i.e., no predefined outputs.  
- **Goal:** Discover hidden patterns, structures, or groupings in data.  
- **Techniques:** Clustering, Dimensionality Reduction.  
- **Example:**  
  - Customer segmentation in marketing.  
  - Grouping customers based on purchasing behavior without prior labels.  

---

### 3. Reinforcement Learning (RL)  
- **Definition:** The model learns by **interacting with an environment** and receiving feedback in the form of rewards or penalties.  
- **Goal:** Maximize cumulative reward through trial and error.  
- **Techniques:** Q-Learning, Policy Gradient methods.  
- **Example:**  
  - Training a robot to walk.  
  - The robot gets a reward for moving correctly and a penalty for falling.  

---

### Summary Table  

| Type | Input Data | Goal | Example |
|------|------------|------|---------|
| Supervised Learning | Labeled | Predict known outcomes | House price prediction |
| Unsupervised Learning | Unlabeled | Discover patterns | Customer segmentation |
| Reinforcement Learning | Feedback-based | Learn optimal actions | Robot walking / Game playing |

---

### Conclusion  
Supervised learning is widely used for prediction tasks, unsupervised learning helps uncover hidden structures, while reinforcement learning focuses on decision-making through feedback. Together, they form the foundation of modern ML applications.  

 ## Q3.Define overfitting, underfitting, and the bias-variance tradeoff in machine learning.

Answer-



### 1. Overfitting  
- **Definition:** Overfitting occurs when a model learns the training data **too well**, including noise and outliers.  
- **Problem:** The model shows very high accuracy on training data but **poor generalization** to unseen data.  
- **Example:**  
  - A decision tree that grows too deep and memorizes every data point instead of learning the general pattern.  

---

### 2. Underfitting  
- **Definition:** Underfitting occurs when a model is **too simple** to capture the underlying patterns in the data.  
- **Problem:** The model performs poorly on both training and test data.  
- **Example:**  
  - Using a straight line (linear regression) to predict a dataset that clearly follows a curve.  

---

### 3. Bias-Variance Tradeoff  
- **Bias:** Error due to overly **simplified assumptions** in the model (leads to underfitting).  
- **Variance:** Error due to model’s **sensitivity to training data** (leads to overfitting).  
- **Tradeoff:**  
  - High bias → Low variance but poor performance (underfit).  
  - High variance → Low bias but poor generalization (overfit).  
  - **Goal:** Achieve the right balance between bias and variance for optimal performance.  

---

### Visual Intuition  

- **High Bias (Underfitting):** Model too rigid → misses patterns.  
- **High Variance (Overfitting):** Model too flexible → memorizes noise.  
- **Balanced Model:** Captures true patterns and generalizes well.  

---

### Conclusion  
Overfitting and underfitting are two extreme problems in ML. The bias-variance tradeoff highlights the importance of balancing model complexity to achieve strong predictive performance on unseen data.  

## Q4. What are outliers in a dataset, and list three common techniques for handling them.  

---

### 1. Definition  
- **Outliers** are data points that significantly **differ from the majority** of observations in a dataset.  
- They may occur due to measurement errors, data entry mistakes, or genuine rare events.  
- Outliers can **distort statistical measures** (like mean, variance) and reduce the accuracy of machine learning models.  

---

### 2. Common Techniques for Handling Outliers  

#### (a) **Removal of Outliers**  
- Delete the extreme values if they are caused by **errors** or irrelevant anomalies.  
- Example: Removing negative ages (−5 years) in a dataset of human ages.  

#### (b) **Transformation of Data**  
- Apply mathematical transformations (e.g., log, square root) to **reduce the influence** of extreme values.  
- Example: Using log transformation on highly skewed income data.  

#### (c) **Capping / Winsorization**  
- Replace extreme outliers with a defined **minimum or maximum threshold** (e.g., 1st and 99th percentile values).  
- Example: Capping extremely high house prices in a real estate dataset.  

---

### 3. Conclusion  
Outliers must be handled carefully: sometimes they contain **important insights** (e.g., fraud detection), while other times they are **errors** that should be removed. The choice of technique depends on the problem context and dataset characteristics.  

 ## Q5. Explain the process of handling missing values and mention one imputation technique for numerical and one for categorical data.

 Answer-
In real-world datasets, missing values are common due to **data entry errors, sensor failures, or incomplete surveys**.  If not handled properly, they may lead to **biased results** and reduce the accuracy of machine learning models.  

---

### 2. Process of Handling Missing Values  

1. **Identification**  
   - Detect missing values using functions such as `isnull()` in Pandas.  

2. **Analysis**  
   - Check the percentage of missing data.  
   - Decide whether to impute or remove based on the extent and importance of the data.  

3. **Treatment**  
   - **Deletion:** Remove rows/columns with excessive missing values.  
   - **Imputation:** Replace missing values with estimated ones.  

---

### 3. Imputation Techniques  

- **For Numerical Data:**  
  - *Mean/Median Imputation* → Replace missing values with the mean or median of the column.  
  - Example: Filling missing student test scores with the class median.  

- **For Categorical Data:**  
  - *Mode Imputation* → Replace missing values with the most frequent category.  
  - Example: Filling missing entries in "Gender" with the most common value (e.g., "Male" or "Female").  

---

### 4. Conclusion  
Handling missing values is a **crucial preprocessing step** in machine learning. The choice of imputation depends on the **nature of data** and whether missingness is random or systematic.

## Q8: Write a Python script to:
- Generate 1000 samples from a normal distribution.
- Introduce 50 random missing values.
- Fill missing values with the column mean.
- Plot a histogram before and after imputation.
(Include your Python code and output in the code box below.)

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Step 1: Generate 1000 samples from normal distribution
np.random.seed(42)  # For reproducibility
data = np.random.normal(loc=50, scale=10, size=1000)  # mean=50, std=10

# Convert to DataFrame
df = pd.DataFrame(data, columns=['Value'])

# Step 2: Introduce 50 random missing values
missing_indices = np.random.choice(df.index, size=50, replace=False)
df.loc[missing_indices, 'Value'] = np.nan

# Show missing values before imputation
print("Missing values before imputation:")
print(df.isna().sum())

# Step 3: Fill missing values with column mean (safe way - no warning)
mean_value = df['Value'].mean()
df_filled = df.copy()
df_filled['Value'] = df_filled['Value'].fillna(mean_value)

# Show missing values after imputation
print("\nMissing values after imputation:")
print(df_filled.isna().sum())

# Step 4: Plot histograms before and after imputation
plt.figure(figsize=(12,5))

# Before imputation
plt.subplot(1,2,1)
df['Value'].hist(bins=30, color='skyblue', edgecolor='black')
plt.title("Before Imputation (with NaNs)")
plt.xlabel("Value")
plt.ylabel("Frequency")

# After imputation
plt.subplot(1,2,2)
df_filled['Value'].hist(bins=30, color='lightgreen', edgecolor='black')
plt.title("After Imputation (mean filled)")
plt.xlabel("Value")
plt.ylabel("Frequency")

plt.tight_layout()
plt.show();


```

## Q9. Implement Min-Max scaling on the following list of numbers [2, 5, 10, 15, 20] using `sklearn.preprocessing.MinMaxScaler`.  
#Print the scaled array. (Include your Python code and output in the code box below.

Answers:-
Min-Max scaling is a normalization technique that transforms values to a fixed range, usually [0, 1].  
The formula is:

$$
X_{\text{scaled}} = \frac{X - X_{\min}}{X_{\max} - X_{\min}}
$$

This ensures the smallest value becomes **0** and the largest becomes **1**, with other values scaled proportionally.


```python



from sklearn.preprocessing import MinMaxScaler
import numpy as np

# Data
data = np.array([[2], [5], [10], [15], [20]])

# Initialize scaler
scaler = MinMaxScaler()

# Fit and transform
scaled_data = scaler.fit_transform(data)

# Print result
print("Original Data:\n", data.ravel())
print("Scaled Data:\n", scaled_data.ravel())
```

## Q10. You are working as a data scientist for a retail company. You receive a customer transaction dataset that contains:
- Missing ages,
- Outliers in transaction amount,
- A highly imbalanced target (fraud vs. non-fraud),
- Categorical variables like payment method.

Explain the step-by-step data preparation plan you’d follow before training a machine learning model. Include how you’d address missing data, outliers, imbalance, and encoding. (Include your Python code and output in the code box below.)

Answer-
**Context:** dataset contains missing `Age`, outliers in `TransactionAmount`, a highly imbalanced target (`Fraud` vs `Non-Fraud`), and categorical variables such as `PaymentMethod`.

---

## 1. Quick data audit (first pass)
- Inspect shape, dtypes, null counts, unique values and basic statistics: `df.info()`, `df.describe()`, `df.isna().sum()`.
- Visual checks: histograms for numeric features, bar plots for categorical distributions, and boxplots for `TransactionAmount`.

**Why:** find scope of problems and inform the next steps (e.g., % missing ages, magnitude of outliers, cardinality of categorical vars).

---

## 2. Split early — preserve a clean test set
- Split dataset into **train** / **validation** / **test** (e.g., 60/20/20) with **stratification** on the fraud label.
- All preprocessing decisions (imputation, scaling, resampling) must be learned only from the **train** set to avoid leakage.

**Why:** ensures honest evaluation and avoids overly optimistic results.

---

## 3. Handle missing values (Age)
- **Exploratory:** Check if missingness is random (MAR/MNAR). Add a boolean indicator column `Age_missing` to signal missingness.
- **Simple approach:** median imputation (robust to outliers) `Age_imputed = Age.fillna(Age.median())`.
- **Better approach:** model-based imputation (KNNImputer or a small regression using other customer features) if relationships exist.
- Always fit imputers on **train** only and transform validation/test.

---

## 4. Treat outliers (TransactionAmount)
- **Identify** outliers using IQR or percentile rules (e.g., below Q1 - 1.5*IQR or above Q3 + 1.5*IQR) or inspect z-scores.
- **Options:**
  - **Transform:** `log(1 + amount)` or Box–Cox to reduce skew and stabilize variance.
  - **Winsorize / Cap:** truncate extremes to 1st/99th percentiles.
  - **Model-aware:** keep extremes if they are meaningful (fraud often occurs in extremes); consider separate flag features for "very large transactions".
- Always apply transformations learned from **train** to other folds.

---

## 5. Feature engineering
- Create time-related features (hour, day-of-week), customer-level aggregates (avg spend, txn count), and missingness indicators.
- Add boolean flags for extreme amounts (e.g., `is_large_amount`) — may be predictive for fraud.

---

## 6. Encode categorical variables
- **Low-cardinality (PaymentMethod):** One-hot encoding (`OneHotEncoder(sparse=False, handle_unknown='ignore')`).
- **High-cardinality:** Target encoding / frequency encoding / leave-one-out target encoding with cross-validation to avoid leakage.
- Use `ColumnTransformer` to apply different transformations to different columns.

---

## 7. Scale numeric features
- For many models (logistic regression, SVM, NN) standardize (`StandardScaler`) or scale to [0,1] (`MinMaxScaler`) after imputation/transform.
- Tree-based models do not require scaling.

---

## 8. Handle class imbalance
- **Prefer to resample only on train set.** Options:
  - **Algorithmic:** use class weights in model (e.g., `class_weight='balanced'` in sklearn).
  - **Resampling:** SMOTE / ADASYN (oversample minority) or combine SMOTE + Tomek (cleaning), or random undersampling of majority.
  - **Ensembles:** use XGBoost/LightGBM with scale_pos_weight or balanced objective.
- Evaluate with PR-AUC, recall, precision, and F1, not just accuracy.

---

## 9. Build pipelines and cross-validation
- Construct `Pipeline` or `ColumnTransformer` that includes imputation, encoding, scaling and (optionally) resampling (use `imblearn.pipeline.Pipeline` for SMOTE).
- Use **StratifiedKFold** for CV to preserve class proportions.
- Use nested CV for hyperparameter tuning if possible.

---

## 10. Final checks, thresholding & monitoring
- Calibrate probabilities (Platt / isotonic) if probabilities are used for decision-making.
- Choose a decision threshold based on business metric (e.g., maximize F1 or meet minimum recall).
- Monitor model drift and performance post-deployment.

---

**Conclusion:** Following the steps above in order (split → impute → outlier handling/transform → encode/scale → resample/train → validate). Prefering pipeline-based code to prevent leakage and ensure reproducibility.


```python


import numpy as np
import pandas as pd

from sklearn.model_selection import train_test_split
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import FunctionTransformer, StandardScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, confusion_matrix

# -----------------------------
# 1) Create an example dataset
# -----------------------------
rng = np.random.default_rng(0)
n = 1000

ages = rng.normal(40, 12, size=n)
missing_age_idx = rng.choice(n, size=int(0.10 * n), replace=False)
ages[missing_age_idx] = np.nan

amounts = rng.exponential(scale=120, size=n)
outlier_idx = rng.choice(n, size=10, replace=False)
amounts[outlier_idx] *= 50

payment_methods = rng.choice(['Card', 'Cash', 'UPI', 'NetBanking'], size=n, p=[0.6, 0.2, 0.15, 0.05])

target = rng.choice([0, 1], size=n, p=[0.97, 0.03])

df = pd.DataFrame({
    'Age': ages,
    'TransactionAmount': amounts,
    'PaymentMethod': payment_methods,
    'Target': target
})

print("Overall class distribution:\n", df['Target'].value_counts(normalize=False), "\n")

# -----------------------------
# 2) Split data
# -----------------------------
train, temp = train_test_split(df, test_size=0.4, stratify=df['Target'], random_state=42)
valid, test = train_test_split(temp, test_size=0.5, stratify=temp['Target'], random_state=42)

# -----------------------------
# 3) Outlier capping
# -----------------------------
low_q, high_q = train['TransactionAmount'].quantile([0.01, 0.99]).values

def cap_amount(series, low, high):
    return series.clip(lower=low, upper=high)

train['TransactionAmount'] = cap_amount(train['TransactionAmount'], low_q, high_q)
valid['TransactionAmount'] = cap_amount(valid['TransactionAmount'], low_q, high_q)
test['TransactionAmount']  = cap_amount(test['TransactionAmount'], low_q, high_q)

# -----------------------------
# 4) Preprocessing pipelines
# -----------------------------
log_transformer = FunctionTransformer(np.log1p, validate=False)

age_pipeline = Pipeline([
    ('imputer', SimpleImputer(strategy='median')),
    ('scaler', StandardScaler())
])

amount_pipeline = Pipeline([
    ('imputer', SimpleImputer(strategy='median')),
    ('log', log_transformer),
    ('scaler', StandardScaler())
])

cat_pipeline = Pipeline([
    ('imputer', SimpleImputer(strategy='most_frequent')),
    ('onehot', OneHotEncoder(handle_unknown='ignore', sparse_output=False))
])

preprocessor = ColumnTransformer(transformers=[
    ('age', age_pipeline, ['Age']),
    ('amt', amount_pipeline, ['TransactionAmount']),
    ('cat', cat_pipeline, ['PaymentMethod'])
])

# -----------------------------
# 5) Classifier
# -----------------------------
clf = Pipeline([
    ('preproc', preprocessor),
    ('clf', LogisticRegression(max_iter=1000, class_weight='balanced', solver='lbfgs'))
])

# -----------------------------
# 6) Train & evaluate
# -----------------------------
X_train = train.drop(columns=['Target'])
y_train = train['Target']
X_valid = valid.drop(columns=['Target'])
y_valid = valid['Target']

clf.fit(X_train, y_train)
y_pred = clf.predict(X_valid)

print("Validation classification report:\n", classification_report(y_valid, y_pred, digits=4))
print("Confusion matrix:\n", confusion_matrix(y_valid, y_pred))

# -----------------------------
# 7) Shape check
# -----------------------------
Xt_train = clf.named_steps['preproc'].transform(X_train)
print("\nTransformed training feature matrix shape:", Xt_train.shape)






```
